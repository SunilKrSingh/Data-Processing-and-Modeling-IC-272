```python
for i in range(len(Xs)):
    print("no of missing values in ", Xs[i], " is: ", df[Xs[i]].isnull().sum())
```

```python
def count4n(df, n):
    count=0
    for i in range(len(df)):
        if df.iloc[i].isnull().sum(0) == n:
            count+=1
    return count
```

```python
C=list()
for i in range(1,13):
    c=count4n(df,i)
    print("no of rows with ", i, " empty values is: ", c)
    C.append(c)
```

```python
plt.plot(range(1,13), C)
```

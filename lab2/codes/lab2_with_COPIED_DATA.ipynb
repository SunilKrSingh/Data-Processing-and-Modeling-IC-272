```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../files/winequality-red_miss-COPY.csv")
```

```python
df.head()
```

```python
cols = [i for i in df]
```

```python
def emptyCount():
    totalEmpty = 0
    for col in cols:
        cnt = df[col].isnull().sum()
        print(col, ": ", cnt)
        totalEmpty += cnt
    print("\nTotal empty cells: ",totalEmpty)
```

```python
df = pd.read_csv("winequality-red.csv", sep=";")
```

```python
cols = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol", "quality"]
```

```python
df.head()
```

```python
#Q1
for col in cols:
    print(col, " : ")
    print("Mean: ",df[col].mean())
    print("Median: ", df[col].median())
    print("Mode: ", df[col].mode())
    print("Min: ", df[col].min())
    print("Max: ", df[col].max())
    print("Std dev: ", df[col].std())
    print()
```

```python
import matplotlib.pyplot as plt
```

```python
for col in cols:
    plt.scatter(df[col], df['quality'])
    plt.legend()
    plt.show()
```

```python
print("Pearson Correlation: ")
for col in cols:
    print(col, ": ", df[col].corr(df['quality']))
```

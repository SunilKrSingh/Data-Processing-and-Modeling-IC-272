```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./pima-indians-diabetes.csv")
```

```python
df.head()
```

```python
cols = ['Age', 'BMI']
```

```python
cols
```

```python
for col in cols:
    print(col,":")
    print("Mean: ",df[col].mean())
    print("Median: ",df[col].median())
    print("Mode: ",float(df[col].mode()))
    print("Min: ",df[col].min())
    print("Max: ",df[col].max())
    print("Std. Dev: ",df[col].std())
    print()
```

```python
plt.scatter(df[cols[0]], df[cols[1]])
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

```python
print("Pearson Correlation Coefficient:")
print(df['Age'].corr(df['BMI']))
```

```python
for col in cols:
    plt.hist(df[col])
    plt.title(col)
    plt.show()
```

```python
grp = list(df.groupby('class'))
```

```python
grp = [grp[i][0] for i in range(len(grp))]
```

```python
grp
```

```python
plt.hist(df['pregs'], bins = grp)
plt.title("pregs grouped by class")
plt.xlabel("class")
plt.ylabel("pregs")
plt.xlim([0,5])
plt.show()
```

```python
for col in cols:
    plt.boxplot(df[col])
    plt.title(col)
    plt.legend()
    plt.show()
```

```python
allCols = [i for i in df]
allCols
```

```python
for i in range(len(allCols)):
    for j in range(len(allCols)):
        print(allCols[i], " vs. ", allCols[j])
        plt.scatter(df[allCols[i]], df[allCols[j]])
        plt.show()
```
